```python
import pandas as pd
import json
import os

# Load CSV into DataFrame
csv_path = "/home/shiv/Documents/exxact/Projects/boltz/Selected_VAV1.csv"
df = pd.read_csv(csv_path)

# Base path for prediction folders
base_dir = "/home/shiv/Documents/exxact/Projects/boltz/boltz_results_VAV1_20250807/predictions"

# Function to get affinity_pred_value for a given compound ID
def get_affinity(compound_id):
    try:
        json_path = os.path.join(base_dir, compound_id, f"affinity_{compound_id}.json")
        with open(json_path, 'r') as f:
            data = json.load(f)
        return data.get("affinity_pred_value", None)
    except Exception as e:
        return None  # or return str(e) for debugging

# Apply function to ID column and add new column
df['affinity_pred_value'] = df['ID'].apply(get_affinity)
df.head(30)
```

```python
import matplotlib.pyplot as plt

# Assuming your dataframe is called df
plt.figure(figsize=(8, 6))
plt.scatter(df["pDC50_nM"], df["affinity_pred_value"], alpha=0.7, edgecolors='k')
plt.ylabel("Affinity Predicted Value")
plt.xlabel("pDC50 (nM)")
plt.title("Scatter Plot: Affinity vs pDC50")
plt.grid(True)
plt.show()
```
